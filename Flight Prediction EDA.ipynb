## Flight Price Prediction[EDA + Feature Engineering]

Q1   What is EDA ?

Ans1 : Exploratory data analysis is the analysis of the data and brings out insights.
    
    It’s storytelling, a story that data is trying to tell. EDA is an approach to analyzing 
    
    the data with the help of various tools and graphical techniques like barplot, histogram, etc.
    
    “Procedures for analyzing data, techniques for interpreting the results of such procedures,
    
    ways of planning the gathering of data to make its analysis easier, more precise or more accurate,
    
    and all the machinery and results of (mathematical) statistics which apply to analyzing data.”
    
    
Q2. What is Feature Engineering?

Ans2: Machine learning fits mathematical notations to the data in order to derive some insights. The models take
    
    features as input. A feature is generally a numeric representation of an aspect ofreal-world phenomena or data.
    
    Just the way there are dead ends in a maze, the path of data is filledwith noise and missing pieces. Our job as 
    
    a Data Scientist is to find a clear path to the end goal of insights.Mathematical formulas work on numerical 
    
    quantities, and raw data isn't exactly numerical. Feature Engineering isthe way of extracting features from data and
    
    transforming them into formats that are suitable for Machine Learning algorithms.

# Feature Engineering on Date_of_Journey

```python
final_df.info()
```

## Creating a Derived Feature
 From Date_of_Journey Feature we will create three fields i.e date,month,year

```python
## Feature Engineering Process

final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

# Converting the Datatype of Date,Month,Year into integer

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

 Dropping the Date_of_Journey feature

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df.head(1)
```

## Feature Engineering on Arrival_Time field

```python
final_df
```

** In Arrival_Time feature we just require the time we don't require the date

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

Different method for doing the same task

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
final_df
```

# Creating a Derived Feature 'Arrival_hour' and 'arrival_min' from Arrival_Time feature

```python


final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head(1)
```

# Converting the datatypes of 'Arrival_hour' and 'Arrival_min' into integer

```python
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
```

```python
final_df.info()
```

# Dropping the Arrival_Time feature

```python
final_df.drop('Arrival_Time',axis=1,inplace=True)
final_df.head(1)
```

## Label Encoding

Label encoding is a simple and straight forward approach. This converts each value in a categorical column
into a numerical value. Each value in a categorical column is called Label.

Label encoding: Assign a unique integer to each label based on alphabetical order

# Label Encoding for Airline,Source,Destination,Additional_Info Features

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
LabelEncoder=LabelEncoder()
```

```python
final_df['Airline']=LabelEncoder.fit_transform(final_df['Airline'])
final_df['Source']=LabelEncoder.fit_transform(final_df['Source'])
final_df['Destination']=LabelEncoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']=LabelEncoder.fit_transform(final_df['Additional_Info'])
```
